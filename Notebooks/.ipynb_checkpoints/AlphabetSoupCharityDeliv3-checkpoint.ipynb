```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("../Resources/charity_data.csv")
application_df.head()
```

```python
# Seems like we need all of the columns except EIN
application_df = application_df.drop(columns=['EIN'])
```

```python
# Set up APPLICATION_TYPE value counts for binning
app_counts = application_df['APPLICATION_TYPE'].value_counts()
app_counts
# Determine which values to replace if counts are less than 500
replace_application = list(app_counts[app_counts < 500].index)
# Replace in dataframe
for app in replace_application:
    application_df.APPLICATION_TYPE = application_df.APPLICATION_TYPE.replace(app,"Other")
    
# Check to make sure binning was successful
application_df.APPLICATION_TYPE.value_counts()
```

```python
# Set up CLASSIFICATION value counts for binning
classification_counts = application_df['CLASSIFICATION'].value_counts()
classification_counts
# Determine which values to replace if counts are less than 1000
replace_class = list(classification_counts[classification_counts < 1000].index)
# Replace in dataframe
for cls in replace_class:
    application_df.CLASSIFICATION = application_df.CLASSIFICATION.replace(cls,"Other")
    
# Check to make sure binning was successful
application_df.CLASSIFICATION.value_counts()
```

```python
# Look at NAME value counts for binning
name_counts = application_df['NAME'].value_counts()
name_counts
```

```python
# Visualize the value counts of NAME
name_counts.plot.density()
```

```python
# I also tried > 5 and > 2 here, and altered every cell below this. They all performed about the same.
name_counts[name_counts > 10]
```

```python
application_df.head()
```

```python
# Determine which values to replace if counts are less than 6
# I also tried < 6 and < 3 here, and altered every cell below this. They all performed about the same.
replace_class = list(name_counts[name_counts < 11].index)
# Replace in dataframe
for cls in replace_class:
    application_df.NAME = application_df.NAME.replace(cls,"Other")
    
# Check to make sure binning was successful
application_df.NAME.value_counts()
```

```python
# Generate our categorical variable lists
application_cat = application_df.dtypes[application_df.dtypes == "object"].index.tolist()
# Create a OneHotEncoder instance
enc = OneHotEncoder(sparse=False)

# Fit and transform the OneHotEncoder using the categorical variable list
encode_df = pd.DataFrame(enc.fit_transform(application_df[application_cat]))
# Add the encoded variable names to the dataframe
encode_df.columns = enc.get_feature_names(application_cat)
encode_df.head()
```

```python
# Merge one-hot encoded features and drop the originals
application_df = application_df.merge(encode_df,left_index=True, right_index=True)
application_df = application_df.drop(application_cat,1)
application_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = application_df["IS_SUCCESSFUL"].values
X = application_df.drop(["IS_SUCCESSFUL"],1).values
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Note that the other groupings of NAME mentioned above produced larger numbers of input features
# and in those cases, the number of nodes in the hidden layers were dialed up appropriately, but it 
# did not improve performance any better than the current implementation.
number_input_features = len(X_train[0])
print(number_input_features)
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len(X_train[0])
hidden_nodes_layer1 = 124
hidden_nodes_layer2 = 62
hidden_nodes_layer3 = 31

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu")
)

# Trying dropout layers
# nn.add(tf.keras.layers.Dropout(0.5))

# Second hidden layer
# nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))

# Trying dropout layers
# nn.add(tf.keras.layers.Dropout(0.5))

# Third hidden layer
# nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer3, activation="relu"))

# Trying dropout layers
# nn.add(tf.keras.layers.Dropout(0.5))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Create a callback that saves the model's weights every 5 epochs.
# Import checkpoint dependencies
import os
from tensorflow.keras.callbacks import ModelCheckpoint

# Define the checkpoint path and filenames
os.makedirs("checkpoints3/",exist_ok=True)
checkpoint_path = "checkpoints3/weights.{epoch:02d}.hdf5"
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
# Create a callback that saves the model's weights every fifth epoch
cp_callback = ModelCheckpoint(
    filepath=checkpoint_path,
    verbose=1,
    save_weights_only=True,
    save_freq='epoch',
    period=5)
```

```python
# Train the model
fit_model = nn.fit(X_train, y_train, epochs=50, callbacks=[cp_callback])
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# I have a suspicion that the accuracy achieved above was a random (in this case good) occurrence.
# Retraining would be unlikely to produce better results and could possibly produce much poorer results.
# All in all, at least 30 different models were trained with various layer configurations and feature counts.
# In all cases, repeated trainings tended to produce accuracies in the range of 0.55 to 0.76. 
# Adding Drop layers increased random training behavior sometimes producing results close
# to the one shown above. Fewer hidden layers (eventually dropping down to one layer) also seemed
# to increase randomness, and the above result is an example of this.
```
